```python
from langchain_community.llms import Ollama

from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory


# 初始化本地模型
llm = Ollama(model="deepseek-r1:14b",
             temperature=0.5
             )  # 指定你在 Ollama 中拉取的模型名称
# 导入对话链和对话内存


# 控制对话系统  保留上下文记忆 用于多轮对话

# 创建对话内存，用于保存历史
memory = ConversationBufferMemory()

# 创建对话链
conversation = ConversationChain(llm=llm, memory=memory)


```

```python
# Prompt 
from langchain.prompts import ChatPromptTemplate
from langchain_core.messages import SystemMessage, HumanMessage

template = ChatPromptTemplate.from_messages([
    SystemMessage(content="""你是一个做菜小助手，名字叫智小厨，可以回答用户的问题。你可以基于以下菜谱知识回答用户的问题
                  菜谱知识：{query}
                  你需要尽量用温柔的语气回答用户的问题。
                  如果根据菜谱知识你不知道答案，可以说“智小厨暂时不知道诶”。
                  你可能需要润色知识库的内容，但不能改变其含义。
                  """),
    HumanMessage(content="用户提问：{question}")
])

prompt = template.format(question="红烧肉怎么做？", query="红烧肉 用料：猪肉，大葱，姜，蒜，八角，桂皮，花椒，料酒，生抽，老抽，白糖，盐，水，油。步骤：1. 猪肉切块，用开水焯水，捞出沥干水分。2. 热锅凉油，放入猪肉煎至两面金黄。3. 放入葱姜蒜煸炒出香味，放入八角桂皮花椒炒出香味。4. 放入料酒，生抽，老抽，白糖，盐，炒匀。5. 加入水没过猪肉，大火烧开，转小火焖煮1小时。6. 焖煮至汤汁浓稠，收汁即可。")
for chunk in llm.stream(prompt):
    print(chunk, end="", flush=True)
# 输出: [SystemMessage(content='你是一个知识渊博的助手，


```
